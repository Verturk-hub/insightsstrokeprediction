```python
import pandas as pd
df=pd.read_csv('healthcare-dataset-stroke-data.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig,axes = plt.subplots(4,2,figsize = (16,16))
sns.set_style('whitegrid')
fig.suptitle("Count plot for various categorical features")

sns.countplot(ax=axes[0,0],data=df,x='gender')
sns.countplot(ax=axes[0,1],data=df,x='hypertension')
sns.countplot(ax=axes[1,0],data=df,x='heart_disease')
sns.countplot(ax=axes[1,1],data=df,x='ever_married')
sns.countplot(ax=axes[2,0],data=df,x='work_type')
sns.countplot(ax=axes[2,1],data=df,x='Residence_type')
sns.countplot(ax=axes[3,0],data=df,x='smoking_status')
sns.countplot(ax=axes[3,1],data=df,x='stroke')
```

* Females are more than males, ever_married ones are more than singles.

* Slightly more people lives in Urban area than Rural.

* There are 5 distinct categories on work type.

* There are 4 different categories on smoking status.

* Stroke data is highly unbalanced.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df.boxplot(column='avg_glucose_level')
plt.title('Box Plot Avg Glucose Level')
plt.show
```

```python
sns.displot(data = df,x = 'age',y = 'bmi',hue = 'stroke')
```

* Stroke is happening mostly after the age of 40.

```python
sns.displot(data = df,x = 'age',y = 'avg_glucose_level',hue = 'stroke')
```

* Average glucose level doesn't effect the stroke.

```python
sns.kdeplot(data = df,x = 'bmi',y = 'age',hue = 'stroke',fill = False)
```

* BMI is not a good indicator of stroke.

```python
sns.displot(x = df.stroke,y = df.age,hue = df.work_type,kind = 'kde')
```

* Private jobs are more prone to stroke while Goverment jobs are less.

```python
boxplot_cols = ['age','avg_glucose_level', 'bmi']
for i, column in enumerate(boxplot_cols):
    sns.boxplot(x='stroke', y=column, data=df)
    plt.show()
```

Bmi doesn't effect stroke.Stroke occurs more when avg. glucose level is higher.Stroke happens mostly in between ages of 60-80.

```python
labels =df['stroke'].value_counts(sort = True).index
sizes = df['stroke'].value_counts(sort = True)

colors = ["lightblue","orange"]
explode = (0.05,0) 
 
plt.figure(figsize=(7,7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90,)

plt.title('Number of stroke in the dataset')
plt.show()
```

Data is highly unbalanced. Stroke event occurs less than 5%.

```python
plt.figure(figsize=(10,5))
placement = 1

for i in ['heart_disease']:
    label = []
    value = []
    for j in range(len(df[i].value_counts().index)):
        label.append(df[i].value_counts().index[j])
        value.append(df[i].value_counts()[j])
        
    plt.subplot(1,2,placement)    
    explode = (0.1, 0.2)
    plt.pie(value, labels = label,autopct='%1.2f%%',colors=['lightblue','orange'],shadow=True,explode=explode)
    plt.title("Heart Disease".format(label[0],label[1]))
    placement += 1
    
plt.tight_layout(pad=0.4)    
```

Heart disease happens slighlty more than stroke.

```python
plt.figure(figsize=(12,8))
placement = 1

for i in ['hypertension']:
    label = []
    value = []
    for j in range(len(df[i].value_counts().index)):
        label.append(df[i].value_counts().index[j])
        value.append(df[i].value_counts()[j])
        
    plt.subplot(1,2,placement)    
    explode = (0.1, 0.2)
    plt.pie(value, labels = label,autopct='%1.2f%%',colors=['lightblue','orange'],shadow=True,explode=explode)
    plt.title("Hypertension".format(label[0],label[1]))
    placement += 1
    
plt.tight_layout(pad=0.4)    
```

Stroke, Hypertension and Heart Disease percents are very low in the dataset. Having the mean age; 43 might be the reason.
